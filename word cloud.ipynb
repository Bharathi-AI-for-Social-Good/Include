```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords
nltk.download('stopwords')

# Load your data
df = pd.read_excel("/content/drive/My Drive/MoE/My Data.xlsx", sheet_name="Sheet1")

# Define non-inclusive and inclusive word mappings
custom_non_inclusive = [
    "chairman", "the handicapped", "afflicted by", "suffers from", "victim of", "fireman",
    "confined to a wheelchair", "wheelchair-bound", "mentally handicapped", "lower class", "non-white"
    "mentally defective", "retarded", "subnormal", "cripple", "invalid", "spastic",
    "able-bodied", "mental patient", "insane", "mad", "deaf and dumb", "deaf mute",
    "the blind", "an epileptic", "diabetic", "depressive", "dwarf", "midget"
]

custom_inclusive = [
    "chairperson", "disabled people", "has a condition", "wheelchair user", "firefighter", "working-class"
    "with a learning disability", "with learning disabilities", "disabled person",
    "person with cerebral palsy", "non-disabled", "person with a mental health condition",
    "hearing impairment", "blind people", "people with visual impairments",
    "person with epilepsy", "diabetes", "depression", "short stature", "restricted growth"
]

# Join all non-inclusive and inclusive text
non_inclusive_text = " ".join(df['non-inclusive'].astype(str).tolist())
inclusive_text = " ".join(df['inclusive'].astype(str).tolist())

# Append custom vocabulary to increase visibility (repeat words to boost frequency)
non_inclusive_text += " " + " ".join(custom_non_inclusive * 5)
inclusive_text += " " + " ".join(custom_inclusive * 5)

# Combine NLTK and custom stopwords
nltk_stopwords = set(stopwords.words('english'))
custom_stopwords = set([
    "just", "let", "need", "would", "like", "always", "because", "us", "also", "make", "even", "get"
])
all_stopwords = nltk_stopwords.union(custom_stopwords)

# Generate word clouds
wordcloud_non = WordCloud(width=800, height=400, background_color='white',
                          stopwords=all_stopwords, colormap='Reds').generate(non_inclusive_text)

wordcloud_incl = WordCloud(width=800, height=400, background_color='white',
                           stopwords=all_stopwords, colormap='Greens').generate(inclusive_text)

# Plot
fig, axs = plt.subplots(1, 2, figsize=(18, 8))

axs[0].imshow(wordcloud_non, interpolation='bilinear')
axs[0].set_title("Non-Inclusive Vocabulary", fontsize=16)
axs[0].axis('off')

axs[1].imshow(wordcloud_incl, interpolation='bilinear')
axs[1].set_title("Inclusive Vocabulary", fontsize=16)
axs[1].axis('off')

plt.tight_layout()
plt.show()
```
